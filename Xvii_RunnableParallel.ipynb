```python
from langchain.schema.runnable import RunnableSequence,RunnableParallel
from langchain.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
prompt1=PromptTemplate(
    template='Give me the details of the topic {topic}\n',
    input_variables=['topic']
)
prompt2=PromptTemplate(
    template='Create a test on the {topic}\n',
    input_variables=['topic']
)
parser=StrOutputParser()
chain=RunnableParallel({
    'details':RunnableSequence(prompt1|llm|parser),
    'test':RunnableSequence(prompt2|llm|parser) }
)
result=chain.invoke('Adolf Hitler')
print(result['details'])
print(result['test'])
```
